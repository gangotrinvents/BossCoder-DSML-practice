Compute the Least Common Multiple (LCM). Create a program to calculate the LCM of two given numbers.


```python
class LCM:
	def __init__(self, a, b):
		self.a = a
		self.b = b

	def gcd(self, x, y):
		
		self.x = x
		self.y = y

		while self.y>0:
			rem = self.x%self.y
			self.x = self.y
			self.y = rem
		return self.x
	def solve(self):
		if self.a > self.b:
			return int((self.a*self.b)/self.gcd(self.a,self.b))
		else:
			return int((self.b*self.a)/self.gcd(self.b,self.a))

l = LCM(6,24)
l.solve()
```

Multiple Inheritance
Q1. Combine two capabilities (fly + swim) in a DroneBoat using multiple inheritance.
Flyer.fly() → returns "flying"
Boat.sail() → returns "sailing"
DroneBoat.move() → return "flying and sailing" using parent methods (no hardcoding strings)

```python
class flyer:
    def fly(self):
        return "flying"
    
class boat:
    def sail(self):
        return "sailing"

class droneboat(flyer,boat):
    def move(self):
        self.fly = flyer.fly(self)
        self.sail = boat.sail(self)

        return f"{self.fly} and {self.sail}"
    
d = droneboat()
d.move()
```

```python
from abc import ABC, abstractmethod
class Payment(ABC):
    @abstractmethod
    def authorize(self):
        pass
    @abstractmethod
    def capture(self):
        pass
class Card(Payment):
    def authorize(self):
        return "Card authorized with card Number"
    def capture(self):
        return "Card payment captured"
class Wallet(Payment):
    def authorize(self):
        return "Wallet authorized"
    def capture(self):
        return "Wallet payment captured"
class NetBanking(Payment):
    def authorize(self):
        return "NetBanking authorized"
    def capture(self):
      return "NetBanking payment captured"
def process(payments):
    for payment in payments:
        print(payment.authorize())
        print(payment.capture())
payments = [Card(), Wallet(), NetBanking()]
process(payments)
```

# Composition
You’re building a Car that needs an Engine, MusicSystem, and GPS. Should you use composition or inheritance? Implement a minimal class layout reflecting your choice.

```python
# Composition will work best here, car is nothing without engine, music and gps. after compisition these will become the part of 
# car and no inheritance required

class engine:
    def strt(self):
        return "engine started"
class music:
    def play(self):
        return "music played"
class gps:
    def track(self):
        return "gps tracking"
class car:
    def __init__(self):
        self.engine = engine()
        self.music = music()
        self.gps = gps()
c = car()
c.engine.strt()
c.music.play()
c.gps.track()
```

```python
class Engine:
    def start(self):
        return "Engine started"
class MusicSystem:
    def play_music(self):
        return "Playing music"
class GPS:
    def navigate(self):
        return "Navigating to destination"
class Car:   # Car HAS-A these components
    def __init__(self):
        self.engine = Engine()
        self.music = MusicSystem()
        self.gps = GPS()
# Example usage
my_car = Car()
print(my_car.engine.start())
print(my_car.music.play_music())
print(my_car.gps.navigate())
```
