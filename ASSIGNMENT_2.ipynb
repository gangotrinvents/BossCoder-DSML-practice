You are given a list of words.
Sort the list using a lambda function based on the length of each word (shortest to longest). Then print the sorted list.
Input Format:
A space-separated list of strings.
Output Format:
A sorted list (based on word length).
Input :
dog, elephant, bat
Output :
'dog', 'bat', 'elephant'


```python
class LambdaSortWords:
    def __init__(self, words):
        self.words = words  # words is a list of strings

    def ss(self,l,a ):
      self.l = l
      self.a = a
      for i in range(len(self.a)):
            for j in range(0,len(self.a)-i-1):
                  if self.a[j] > self.a[j+1]:
                        self.a[j],self.a[j+1] = self.a[j+1],self.a[j]
                        self.l[j],self.l[j+1] = self.l[j+1],self.l[j]
      return self.l,self.a

    def solve(self):
        # Sort the list by length using lambda
        l,a = self.ss(self.words,list(map(lambda x: len(x), self.words)))

        return l
```

Write a program using functools.reduce() and lambda to compute the product of all numbers in a given list.
Input Format:
A single line of space-separated integers.
Output Format:
An integer representing the product of all elements.
Input :
2,3,4
Output :
24

```python
from functools import reduce

class ProductOfAllNumbers:
	def __init__(self, numbers):
		self.numbers = numbers

	def solve(self):
		return reduce(lambda a,b : a*b, self.numbers)
```

Given a sentence, reverse every word in it using map() and lambda.
Return a single space-separated string with each word reversed.
Input Format:
A single line sentence (words separated by space).
Output Format:
A single string with each word reversed.
Input :
hello world
Output :
olleh dlrow

```python
class ReverseWordsWithLambda:
	def __init__(self, sentence):
		self.sentence = sentence

	def solve(self):
		return " ".join(list(map(lambda x : x[::-1], self.sentence.split())))
c = ReverseWordsWithLambda('hello world')

print(c.solve())
```

You are given a list of integers.
Use filter() to select even numbers, then use map() and lambda to square them. Finally, print the sum of these squares.
Input Format:
A single line of integers.
Output Format:
A single integer: the sum of squares of even numbers.
Input :
1, 2, 3, 4, 5
Output :
20

```python
class SumOfSquaresOfEvenNumbers:
	def __init__(self, number_line):
		self.number_line = number_line

	def solve(self):
		return sum(list(map(lambda x : x**2, list(filter(lambda x : x%2==0, self.number_line)))))
s = SumOfSquaresOfEvenNumbers([1, 2, 3, 4, 5])
a = s.solve()
a
```

class ZipAndConcatenate: def __init__(self, list1, list2): self.list1 = list1 self.list2 = list2 def solve(self): result = list(map(lambda x: x[0] + x[1], zip(self.list1, self.list2))) return(result)
Input :
a, b, c
x, y, z
Output :
'ax', 'by', 'cz'

```python
class ZipAndConcatenate:
	def __init__(self, list1, list2):
		self.list1 = list1
		self.list2 = list2

	def solve(self):
		ls = []
		for i,j in zip(self.list1, self.list2):
			ls.append(i+j)
		return ls
```

Determine whether a number is binary. Write a program that checks if a number is binary by converting it to a string and verifying if it contains only '0' and '1' characters. If it contains any other characters, it is not a binary number.



Input: 101001
Output: true

```python
class CheckBinary:
	def __init__(self, s):
		self.s = s

	def check(self):
            x = str(self.s)
            print(x)
            check = 0
            for i in x[:]:
                  print(i)
                  if i == '0' or i == '1':
                        check += 1
                  else:
                        check = -1
                        break
            if check >= 0:
                  return 'true'
            else:
                  return 'false'
            
c = CheckBinary('100110')
d = c.check()
d
```

difference between functional programming and object-oriented programming?

**FUNCTIONAL PROGRAMMING (FP)**
**Philosophy**:

“Don’t change things. Create new things.”

FP rules (core idea)

No mutation

No hidden side effects

Same input → same output (pure)

```python
def add(x, y):
    return x + y
#✔ No state
#✔ No memory change
#✔ Predictable
```

**OBJECT-ORIENTED PROGRAMMING (OOP)**
**Philosophy:**

“Objects hold data and change over time.”

OOP rules (core idea)

Objects own data

Methods modify object state

State evolution matters

```python
class Account:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount   # mutation
```

```python
from functools import reduce

def decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Inputs: {args}")
        result = func(*args, **kwargs)
        print(f"Result: {result}")
        return result
    return wrapper

@decorator
def process_data(data):
    print(data)
    mapped = map(lambda x: x * 2, data)
    print(mapped)
    filtered = filter(lambda x: x % 3 == 0, mapped)
    print(filtered)
    reduced = reduce(lambda x, y: x + y, filtered, 10)
    print(reduced)
    return reduced

input_data = [1, 2, 3, 4, 5, 6]
final_result = process_data(input_data)
print(final_result)
```
